```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

comb = pd.read_csv('lOBFeatures', parse_dates=True, dayfirst=True)
comb.date.unique()
```

```python
comb['date'] = pd.to_datetime(comb['date'], format='%d/%m/%Y')
df = comb.sort_values(by='date')
df.date.unique()
```

```python
jan17 = df[df['date']=='2021-01-17'].sort_values(by='time')[:500]
jan21 = df[df['date']=='2021-01-21'].sort_values(by='time')[-500:]
```

```python
df
```

```python

plt.figure()
plt.subplot(1,2,1)
plt.plot(jan17.time , jan17.MicroPrice)
plt.subplot(1,2,2)
plt.plot(jan21.time , jan21.MicroPrice)
plt.show()
```

```python
plt.plot(jan17.time, jan17.askAvg)
plt.plot(jan17.time, jan17.bidAvg)
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(jan17.time, jan17.Price)
plt.title('Time v Price')
plt.subplot(2,1,2)
# plt.plot(jan17.time, jan17.numberOfAsks)
plt.plot(jan17.time, jan17.numberOfBids)
plt.title('# of Bids')
```

```python
dailySummary =  pd.read_csv('tapesDailySummary.csv', parse_dates=True, dayfirst=True)
dailySummary = dailySummary.rename(columns={'ClosingPrice':'closingPrice', 'HighestPrice':'highestPrice', 'LowestPrice':'lowestPrice','VolumeTraded':'volumeTraded','DaysUntilWeekend':'daysUntilWeekend', 'DaysUntilMonthEnd':'daysUntilMonthEnd'})
df = dailySummary
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df = df.sort_values(by='date')
df['ind']=np.arange(len(df))
```

```python
df
```

```python
# df['SMA10Days'] = df['closingPrice'].rolling(window=10, min_periods=10)
df['EMA10Days'] = df['closingPrice'].ewm(span=10, adjust=False).mean()
df['EMA16Days'] = df['closingPrice'].ewm(span=16, adjust=False).mean()
df['EMA22Days'] = df['closingPrice'].ewm(span=22, adjust=False).mean()
```

```python

df['SMA10Days'] = df['closingPrice'].rolling(window = 10, min_periods=10).mean()
df['SMA16Days'] = df['closingPrice'].rolling(window = 16, min_periods=16).mean()
df['SMA22Days'] = df['closingPrice'].rolling(window = 22, min_periods=22).mean()
```

```python
plt.plot(df.date, df.closingPrice)
```

```python
plt.plot(df.date, df.EMA10Days)

# plt.plot(df.date, df.closingPrice, c = 'black')
plt.plot(df.date, df.EMA16Days)
plt.plot(df.date, df.EMA22Days)
plt.title("EMA Compared")
```
