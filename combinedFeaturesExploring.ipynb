```python
import datetime

import pandas as pd

dailySummary = pd.read_csv('/Users/brookegrantham/Documents/DSMP hm/dailySummaryFinal')
dailySummary
```

```python
del dailySummary['Unnamed: 0']
```

```python
dailySummary
```

```python
from datetime import date
from matplotlib import dates as dates
```

```python
for i in range(len(dailySummary)):
    dailySummary['date'][i]=dailySummary['date'][i].replace('-','/')
    dailySummary['date'][i]= dates.datestr2num(dailySummary['date'][i])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
ax = plt.subplot(2,1,1)
ax.plot(dailySummary.date, dailySummary.SMA10Days, label='10 Days')
ax.plot(dailySummary.date, dailySummary.SMA16Days, label='16 Days')
ax.plot(dailySummary.date, dailySummary.SMA22Days, label='22 Days')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.title('Simple Moving Average (SMA)')
plt.xticks(rotation=45)
ax.xaxis.set_major_formatter(dates.DateFormatter('%d/%m/%Y'))
ax = plt.subplot(2,1,2)
ax.plot(dailySummary.date, dailySummary.EMA10Days, label='10 Days')
ax.plot(dailySummary.date, dailySummary.EMA16Days, label='16 Days')
ax.plot(dailySummary.date, dailySummary.EMA22Days, label='22 Days')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.title('Exponential Moving Average (EMA)')
plt.xticks(rotation=45)
ax.xaxis.set_major_formatter(dates.DateFormatter('%d/%m/%Y'))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)
```
