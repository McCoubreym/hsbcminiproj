```python
combinedfeatures = pd.read_csv('0102CombinedFeatures.csv')
combinedfeatures
```

```python
from matplotlib import pyplot as plt
import numpy as np

_ = plt.scatter(combinedfeatures.time,combinedfeatures.MicroPrice, c='black')
line = np.polyfit(combinedfeatures.time,combinedfeatures.MicroPrice, 1)
p = np.poly1d(line)
plt.plot(combinedfeatures.time,p(combinedfeatures.time))
```
